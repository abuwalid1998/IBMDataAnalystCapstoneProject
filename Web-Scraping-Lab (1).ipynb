## Extract information from the given web site
You will extract the data from the below web site: <br> 


```python
#this url contains the data you need to scrape
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/labs/datasets/Programming_Languages.html"
```

The data you need to scrape is the **name of the programming language** and **average annual salary**.<br> It is a good idea to open the url in your web broswer and study the contents of the web page before you start to scrape.


Import the required libraries


```python
# Your code here

from bs4 import BeautifulSoup # this module helps in web scrapping.
import requests  # this module helps us to download a web page
```

Download the webpage at the url


```python
#your code goes here
data  = requests.get(url).text 
```

Create a soup object


```python
#your code goes here
soup = BeautifulSoup(data,"html5lib")
```

Scrape the `Language name` and `annual average salary`.


```python
#your code goes here
table = soup.find('table')
#Get all rows from the table
for row in table.find_all('tr'): # in html table row is represented by the tag <tr>
    # Get all columns in each row.
    cols = row.find_all('td') # in html a column is represented by the tag <td>
    no = cols[0].getText() 
    language = cols[1].getText() 
    Created_By = cols[2].getText() 
    Average_Annual_Salary = cols[3].getText() 
    Learning_Difficulty = cols[4].getText() 
    print("{},{},{},{},{}".format(no,language,Created_By,Average_Annual_Salary,Learning_Difficulty))
```

Save the scrapped data into a file named *popular-languages.csv*


```python
# your code goes here
import csv

response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    soup = BeautifulSoup(response.content, 'html.parser')
    table = soup.find('table')

    # Open the CSV file for writing
    with open('popular-languages.csv', mode='w', newline='') as csv_file:
        csv_writer = csv.writer(csv_file)

        # Write the header row to the CSV file
        csv_writer.writerow(['No', 'Language', 'Created By', 'Average Annual Salary', 'Learning Difficulty'])

        # Get all rows from the table
        for row in table.find_all('tr'):
            # Get all columns in each row.
            cols = row.find_all('td')

            if len(cols) >= 5:
                no = cols[0].getText()
                language = cols[1].getText()
                created_by = cols[2].getText()
                average_annual_salary = cols[3].getText()
                learning_difficulty = cols[4].getText()

                # Write the data to the CSV file
                csv_writer.writerow([no, language, created_by, average_annual_salary, learning_difficulty])

    print("Data has been saved to popular-languages.csv")
else:
    print(f"Failed to fetch data. Status code: {response.status_code}")
```
